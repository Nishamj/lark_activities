```python
df.drop(['Name','SibSp','Parch','Ticket','Cabin','Embarked','Last_Name','First_Name', 'Title'],axis='columns',inplace=True)
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
inputs.head()
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head()
```

```python
#Dropping male column because of dummy variable trap theory
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
# Gaussian Naive Bayes classifier
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy score: ',accuracy_score (y_pred,y_test))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
cv_method = RepeatedStratifiedKFold(n_splits=5,n_repeats=3,random_state=999)
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.model_selection import GridSearchCV
params_NB = {'var_smoothing': np.logspace(0,-9, num=100)}

gs_NB = GridSearchCV(estimator=model,param_grid=params_NB,cv=cv_method,verbose=1,scoring='accuracy')

Data_transformed = PowerTransformer().fit_transform(X_test)

gs_NB.fit(Data_transformed, y_test);
```

```python
results_NB = pd.DataFrame(gs_NB.cv_results_['params'])
```

```python

results_NB['test_score'] = gs_NB.cv_results_['mean_test_score']
```
