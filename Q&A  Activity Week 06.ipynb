# 2. Examine the data by using visualizations.


```python
plt.figure(figsize=(16,6))
plt.bar(df['Region'],df['GDP per capita (current US$)'],color='green')
plt.title('Region vs GDP per capita')
plt.xticks(rotation=90)
plt.yticks()
plt.xlabel('Region')
plt.ylabel('GDP')
```

```python
plt.figure(figsize=(16,6))
plt.bar(df['Region'],df['Population in thousands (2017)'],color='red')
plt.title('Region vs Population')
plt.xticks(rotation=90)
plt.yticks()
plt.xlabel('Region')
plt.ylabel('Population')
```

```python
plt.figure(figsize=(16,6))
plt.bar(df['Region'],df['International trade: Imports (million US$)'],color='purple')
plt.title('Region vs Imports')
plt.xticks(rotation=90)
plt.yticks()
plt.xlabel('Region')
plt.ylabel('Imports')
```

```python
plt.figure(figsize=(16,6))
plt.bar(df['Region'],df['Surface area (km2)'],color='cyan')
plt.title('Region vs Surface area')
plt.xticks(rotation=90)
plt.yticks()
plt.xlabel('Region')
plt.ylabel('Surface area')
```

# 4. On average which region has the highest
a. GDP per capita (current US$)

b. International trade: Imports (million US$)


```python
GDP=df.groupby('Region')[['GDP per capita (current US$)' ]].mean()
GDP
```

```python
GDP.nlargest(1,'GDP per capita (current US$)')
```

```python
imports=df.groupby('Region')[['International trade: Imports (million US$)' ]].mean()
imports
```

```python
imports.nlargest(1,'International trade: Imports (million US$)')
```
