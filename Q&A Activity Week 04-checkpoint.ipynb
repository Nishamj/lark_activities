# 1.Compare churn count with respect to gender

```python
print('Gender values\n',df['gender'].value_counts())
plt.figure(figsize=(6,5))
sns.countplot(data=df,x='gender',hue='Churn')
```

# 2. Find out how many female senior citizens there in the dataset


```python
femalesc=df[(df['SeniorCitizen']==1) & (df['gender']=='Female')]
femalesc['gender'].value_counts()
```

```python
plt.figure(figsize=(3,6))
sns.countplot(data=df,x=femalesc['gender'])
plt.title('Senior citizens')
```

# 3. Compare 'tenure' with 'Total Charges'


```python
plt.figure(figsize=(10,10))
plt.scatter(df['tenure'],df['TotalCharges'],s=15,color='red',marker='^')
plt.title("Tenure vs Total Charges",fontsize=18)
plt.xlabel('Tenure')
plt.ylabel('Total charges')
```

# 4. Find out which contract is preferred by the senior citizen.


```python
senior_citizen=df[df['SeniorCitizen']==1]
senior_citizen
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x=senior_citizen['Contract'])
plt.title('Contract preferred by Senior citizens')
```
